# Intelligence Briefing Format Analysis

**Project**: CDDBS (Counter-Disinformation Database System)  
**Sprint**: 1 - Briefing Format Redesign  
**Task**: 2.1 - Research Intelligence Briefing Formats  
**Goal**: Analyze 10 professional intelligence briefings and document best practices

---

## Research Objectives

1. Study professional intelligence briefing formats from established organizations
2. Identify structural patterns common to effective briefings
3. Document attribution and sourcing best practices
4. Analyze confidence signaling techniques
5. Propose a CDDBS-specific briefing template

## Analysis Framework

For each briefing analyzed, document the following:

| Dimension | Description |
|-----------|-------------|
| **Source Organization** | Who published the briefing |
| **Title & Date** | Full title and publication date |
| **URL** | Link to the original briefing |
| **Structure** | How the briefing is organized (sections, headings) |
| **Attribution Model** | How claims are attributed to sources |
| **Confidence Signaling** | How certainty/uncertainty is communicated |
| **Evidence Presentation** | How evidence is structured and referenced |
| **Visual Elements** | Use of images, charts, timelines |
| **Actionability** | Whether conclusions lead to clear actions |
| **Strengths** | What works well in this format |
| **Weaknesses** | What could be improved |
| **Relevance to CDDBS** | What we can adopt for our system |

## Briefing 1: EUvsDisinfo

### Source
- **Organization**: EU EastStratCom Task Force / EUvsDisinfo
- **Title**: Disinformation Cases Database
- **Date**: Ongoing (established 2015, 15,000+ cases)
- **URL**: https://euvsdisinfo.eu/disinformation-cases/

### Structure
Each database entry is structured as a **claim-rebuttal pair** with consistent metadata:
1. **Title** - headline of the disinformation claim
2. **Date** - publication date of the disinformation
3. **Outlets** - media outlets that propagated the claim
4. **Country** - target countries affected
5. **Disinformation** - full text of the false claim
6. **Information** - factual rebuttal/disproof with evidence
7. **Keywords/Tags** - categorical labels linking to broader narratives

The format is deliberately structured as a searchable database rather than narrative reports. Each case is atomic — one claim, one rebuttal.

### Attribution Model
- Claims are attributed to specific **outlets** (named media sources like RT, Sputnik, etc.)
- Rebuttals reference trustworthy sources with inline **hyperlinks** to evidence
- Academic analysis of 350 URLs in response sections confirmed all references link to either other fact-checking agencies (Bellingcat, StopFake) or EUvsDisinfo's own prior debunks
- Attribution is **source-level** — the outlet is always named, not just "Russian media"

### Confidence Signaling
- **Minimal explicit confidence signaling** — EUvsDisinfo treats cases as binary (disinformation vs. fact)
- No formal confidence scale or uncertainty language
- The categorization itself implies high confidence — a case only enters the database when the team is certain it meets their criteria
- This binary approach is a deliberate editorial choice suited to a public-facing database

### Evidence Presentation
- Text-based rebuttals with hyperlinks to authoritative sources
- No embedded screenshots or visual evidence in the database entries themselves
- References to primary sources (government statements, official statistics, news reports)
- Relies on the credibility of cited sources rather than presenting raw evidence

### Visual Elements
- Clean, tabular database interface for browsing cases
- Filterable by date, country, outlet, keyword
- No charts, network graphs, or visual analysis within individual entries
- Weekly update reports with aggregate statistics

### Actionability
- High actionability for fact-checkers and journalists — each case is a ready-to-use rebuttal
- Less actionable for policy-makers — lacks recommendations or strategic assessment
- Designed as a **reference tool** rather than an analytical product

### Strengths
- Massive scale (15,000+ cases) provides comprehensive coverage
- Consistent structure makes comparison and pattern analysis possible
- Open-source and searchable — high accessibility
- Multilingual monitoring across EU languages
- Weekly cadence ensures timeliness
- Clear separation of claim vs. rebuttal

### Weaknesses
- No confidence gradations — everything is treated as certain
- No network or behavioral analysis — focuses only on content
- Individual entries lack strategic context (how does this fit broader campaigns?)
- No visual evidence or multimedia
- Database format may overwhelm non-specialist users

### CDDBS Takeaways
- **Adopt**: Consistent metadata structure (date, source, target country, keywords)
- **Adopt**: Clear separation of claim/evidence from analytical assessment
- **Adopt**: Hyperlinked attribution to primary sources
- **Improve on**: Add confidence gradations instead of binary classification
- **Improve on**: Add behavioral/network context alongside content analysis
- **Improve on**: Include visual evidence (screenshots, charts)

## Briefing 2: DFRLab (Atlantic Council)

### Source
- **Organization**: Digital Forensic Research Lab (Atlantic Council)
- **Title**: Various reports on influence operations and digital forensics
- **Date**: Ongoing (est. 2016)
- **URL**: https://www.atlanticcouncil.org/programs/digital-forensic-research-lab/

### Structure
DFRLab reports typically follow this structure:
1. **Headline/Title** - descriptive, often using their "#[Topic]" tagging convention
2. **Key Takeaway / Executive Summary** - 1-2 paragraph overview
3. **Background/Context** - situating the finding in broader campaigns
4. **Methodology** - how the data was collected and analyzed
5. **Detailed Findings** - broken into sub-sections with evidence
6. **Network/Behavioral Analysis** - coordination patterns, bot detection
7. **Conclusion** - summarizing implications
8. **Appendix** - data tables, additional evidence

DFRLab was a key partner in the Election Integrity Partnership (EIP) alongside Stanford IO, Graphika, and the University of Washington CIP, establishing cross-organizational analytical standards.

### Attribution Model
- **Granular attribution**: Claims traced to specific accounts, posts, and timestamps
- Uses **archived screenshots** as primary evidence (preserving deleted content)
- Cross-references multiple platforms (Twitter, Facebook, Instagram, Telegram)
- Attributes behavioral patterns (not just content) — e.g., "this network of accounts exhibited coordinated behavior consistent with..."
- Partners with platforms (Meta, Twitter) who share takedown datasets for independent analysis

### Confidence Signaling
- Uses hedging language: "consistent with," "suggests," "indicates"
- Distinguishes between what is **observed** vs. **assessed**
- Does not use a formal numerical or categorical confidence scale
- Relies on weight of evidence and multiple indicators rather than explicit confidence statements
- Methodology section provides implicit confidence through transparency

### Evidence Presentation
- **Screenshots** of social media posts (often annotated)
- **Network visualization** graphs showing account relationships
- **Timeline charts** showing activity patterns
- **Data tables** with quantitative metrics (posting frequency, engagement rates)
- **CrowdTangle** and platform analytics data
- Side-by-side comparisons of coordinated content

### Visual Elements
- Heavy use of annotated screenshots
- Network graphs (node-link diagrams)
- Bar charts and time-series plots for activity patterns
- Maps for geolocation analysis
- Before/after comparisons of platform takedowns

### Actionability
- Highly actionable for platforms — findings often lead directly to account takedowns
- Actionable for researchers — methodology is reproducible
- Less directly actionable for general public — requires domain expertise to interpret

### Strengths
- Rigorous digital forensics methodology
- Strong visual evidence (screenshots, network graphs)
- Cross-platform analysis capability
- Partnership with platforms provides access to takedown data
- Transparent methodology enables peer review
- Rapid turnaround for emerging events

### Weaknesses
- Reports can be lengthy and technical for non-specialists
- Confidence signaling is implicit rather than explicit
- No standardized severity or priority rating
- Individual reports don't always link to broader strategic picture

### CDDBS Takeaways
- **Adopt**: Archived screenshot evidence (critical for social media analysis)
- **Adopt**: Network visualization showing account relationships
- **Adopt**: Activity timeline charts
- **Adopt**: Cross-platform analysis approach
- **Adopt**: Distinction between observed behavior and analytical assessment
- **Improve on**: Add explicit confidence scale (formalize what DFRLab does implicitly)
- **Improve on**: Add severity/priority rating per finding

## Briefing 3: Bellingcat

### Source
- **Organization**: Bellingcat
- **Title**: Various OSINT investigations (e.g., MH17, Skripal poisoning, Syria chemical weapons)
- **Date**: Ongoing (est. 2014)
- **URL**: https://www.bellingcat.com/

### Structure
Bellingcat investigations follow a **narrative investigative journalism** format:
1. **Headline** - attention-grabbing, specific claim
2. **Lead/Hook** - opening paragraph establishing what was discovered
3. **Background** - context for why this matters
4. **Methodology Narrative** - step-by-step explanation of how the investigation was conducted (the "detective story")
5. **Evidence Presentation** - layered, building from individual pieces to larger conclusions
6. **Verification Steps** - showing how each piece of evidence was verified
7. **Conclusion** - synthesizing findings into clear assessment
8. **Author Bio** - transparency about who conducted the investigation

The structure is deliberately **narrative** — it tells a story of discovery, which makes complex OSINT accessible to general audiences.

### Attribution Model
- **Source-transparent**: Every piece of evidence is traced to its origin (satellite imagery provider, social media platform, public records database)
- Uses **archived/cached versions** of content to prevent link rot
- Cross-references multiple independent open sources
- The Justice & Accountability Unit (J&A Unit, est. 2022 with GLAN) formalized methodology for **legal admissibility** — evidence chains documented to courtroom standards
- Attributes tools used for verification (geolocation services, reverse image search, flight tracking)

### Confidence Signaling
- **Hedged but clear language**: "We can confirm," "Evidence strongly suggests," "We were unable to verify"
- Uses cumulative evidence approach — confidence builds through multiple independent verification steps
- Distinguishes between "verified" (independently confirmed through multiple sources) and "assessed" (analytical judgment)
- Acknowledges gaps explicitly: "We were unable to determine..." or "This could not be independently verified"
- No formal confidence scale, but the investigative narrative implicitly communicates confidence through the weight of evidence shown

### Evidence Presentation
- **Annotated satellite imagery** — before/after comparisons with markup
- **Geolocated photos/videos** — matching visual features to known locations
- **Social media screenshots** with metadata analysis
- **Document analysis** — travel records, phone records, open databases
- **Side-by-side comparisons** — matching faces, objects, locations
- Step-by-step **verification walkthroughs** that readers can replicate

### Visual Elements
- Heavily visual — images are core to the argument, not supplementary
- Annotated satellite imagery with arrows, circles, labels
- Interactive maps and timeline tools
- Side-by-side photo comparisons
- Schematic diagrams explaining complex relationships
- Embedded social media posts (often archived)

### Actionability
- Highly actionable for law enforcement and prosecutors (J&A Unit outputs designed for legal proceedings)
- Actionable for journalists — provides leads and verified facts for further reporting
- Empowering for public — shows methodology that others can learn and replicate
- Less structured for policy recommendations — focused on establishing facts rather than prescribing actions

### Strengths
- Gold standard for OSINT methodology and transparency
- Narrative format makes complex investigations accessible to general audiences
- Legal-grade evidence chain documentation (via J&A Unit)
- Strong visual evidence presentation
- Methodology is educational — teaches readers how to verify
- Community-driven with volunteer contributors
- Track record of investigations leading to real-world accountability

### Weaknesses
- Narrative format is lengthy — not suited for quick reference or automated consumption
- No standardized output format — each investigation is unique
- Doesn't scale to automated analysis (each investigation is bespoke)
- Limited network-level analysis — focuses on individual actors/events

### CDDBS Takeaways
- **Adopt**: Rigorous source attribution with tool transparency
- **Adopt**: Visual evidence as core (not supplementary) to the analysis
- **Adopt**: Clear distinction between "verified" and "assessed" claims
- **Adopt**: Explicit acknowledgment of gaps and limitations
- **Learn from**: Narrative structure makes findings accessible — CDDBS briefings should be readable, not just data dumps
- **Improve on**: Create structured, repeatable format (what Bellingcat does bespoke, CDDBS should systematize)
- **Improve on**: Add formal confidence scale to complement qualitative language

## Briefing 4: NATO StratCom COE

### Source
- **Organization**: NATO Strategic Communications Centre of Excellence (Riga, Latvia)
- **Title**: Various publications + academic journal *Defence Strategic Communications*
- **Date**: Ongoing (est. 2014)
- **URL**: https://stratcomcoe.org/publications

### Structure
NATO StratCom COE produces multiple publication types with distinct structures:

**Research Reports** (20+ per year):
1. **Title Page** - with NATO branding and classification
2. **Executive Summary** - key findings and recommendations
3. **Table of Contents** - formal academic structure
4. **Introduction** - problem statement, scope, objectives
5. **Literature Review / Background** - prior work and context
6. **Methodology** - research design and data collection
7. **Findings** - organized by theme or research question
8. **Analysis & Discussion** - implications of findings
9. **Recommendations** - specific, actionable recommendations for NATO/Allies
10. **References** - full academic bibliography
11. **Appendices** - supplementary data, tools, frameworks

**Academic Journal** (*Defence Strategic Communications*):
- Bi-annual, peer-reviewed, open-access
- Follows standard academic article format (abstract, introduction, methods, results, discussion, references)

**Doctrine Documents** (2024 milestone — first NATO StratCom fundamentals doctrine):
- Formal NATO doctrine format with standardized terminology

### Attribution Model
- **Academic citation standard** — footnotes and bibliography
- References to primary NATO documents, allied government publications
- Cites specific data sources and methodologies
- Peer review process for the academic journal provides additional quality assurance
- Attributes contributions to specific researchers/authors

### Confidence Signaling
- Uses **formal analytical language** consistent with NATO intelligence assessment standards
- Hedging language: "evidence suggests," "assessment indicates," "findings are consistent with"
- Peer review process implicitly validates confidence
- Formal recommendations imply the organization stands behind the assessment
- Some reports use explicit **threat level** or **severity** ratings

### Evidence Presentation
- Quantitative data with charts and statistical analysis
- Case studies with detailed evidence chains
- Social media data analysis with visualizations
- Survey data from NATO member states
- Comparative analysis across countries/platforms

### Visual Elements
- Professional publication design with NATO branding
- Charts, graphs, and data visualizations
- Infographics summarizing key findings
- Maps showing geographic scope
- Tables with comparative data

### Actionability
- **Highly actionable** — reports are designed to inform NATO policy and doctrine
- Recommendations are specific, addressed to identified stakeholders
- Doctrine documents directly shape NATO communication practices
- Training materials support practical implementation

### Strengths
- Institutional authority and NATO backing
- Rigorous academic methodology (peer-reviewed journal)
- Direct policy impact — recommendations lead to doctrinal changes
- Multi-national perspective (12+ sponsoring nations)
- Comprehensive coverage of strategic communications landscape
- First-ever NATO StratCom doctrine published in 2024

### Weaknesses
- Publication cycle can be slow (academic pace vs. real-time threats)
- Focus on strategic/policy level — less tactical/operational detail for individual cases
- NATO-centric perspective may not fully capture non-Western information environments
- Formal tone may limit accessibility for non-specialist audiences

### CDDBS Takeaways
- **Adopt**: Formal recommendation structure (specific, actionable, addressed to stakeholders)
- **Adopt**: Multi-layered analysis (strategic context + tactical evidence)
- **Adopt**: Academic citation standards for attribution
- **Learn from**: Doctrine approach — standardized frameworks and terminology
- **Learn from**: Threat/severity ratings for findings
- **Improve on**: Faster publication cadence (automated analysis enables real-time output)
- **Improve on**: More accessible tone for non-specialist users

- Platform-provided datasets cited with methodology for selection
- Cross-organizational collaboration provides multi-source verification
- Transparent about data access limitations and platform relationships
- Archived social media data with timestamps

### Confidence Signaling
- Academic hedging language with clear assessment statements
- Distinguishes between "observed" data patterns and "assessed" causation
- Methodology transparency allows readers to evaluate confidence
- Multi-organization partnership (EIP) implicitly raises confidence through independent corroboration
- Acknowledges limitations of data access and analytical scope

### Evidence Presentation
- Quantitative social media data analysis (volume, engagement, network metrics)
- Timeline visualizations of narrative spread
- Network graphs showing information flow
- Platform-specific data tables
- Case study narratives with embedded evidence
- Statistical analysis with methodology disclosure

### Visual Elements
- Data visualizations (charts, graphs, network diagrams)
- Timeline plots showing narrative evolution
- Geographic distribution maps
- Platform comparison tables
- Infographics for public-facing summaries

### Actionability
- Highly actionable for platforms — findings shared directly for content moderation
- Actionable for policymakers — specific legislative and regulatory recommendations
- "Data memo" format enables rapid response

### Strengths
- Academic rigor with policy relevance
- Multi-platform, multi-method analysis
- Rapid-response capability ("data memos")
- Cross-organizational collaboration model (EIP)
- Transparent methodology and data access
- Balance of quantitative and qualitative analysis

### Weaknesses
- Institutional vulnerability — political pressure led to organizational scaling back in 2024
- Dependency on platform data access (which can be revoked)
- Some reports dense and academic for non-specialist audiences
- No standardized confidence scale across publications

### CDDBS Takeaways
- **Adopt**: "Data memo" format for rapid-response outputs
- **Adopt**: Multi-method analysis (quantitative + qualitative + case studies)
- **Adopt**: Clear methodology disclosure in every output
- **Adopt**: Platform-specific data tables with metrics
- **Learn from**: The EIP collaboration model (how to standardize across analysts)
- **Improve on**: Add formal confidence scale
- **Improve on**: Build institutional resilience — don't depend on platform data access alone

## Briefing 6: Graphika

### Source
- **Organization**: Graphika
- **Title**: Various reports (e.g., COVID-19 Infodemic, White Helmets network analysis, AI-generated laundering reports)
- **Date**: Ongoing
- **URL**: https://graphika.com/reports

### Structure
Graphika reports follow a **network-analysis-first** structure:
1. **Executive Summary** - key findings with network context
2. **Chronological Mapping Series** - temporal snapshots of network evolution
3. **Map Background** - explanation of the network mapping methodology
4. **Network Analysis** - core section, visualizing community structures and relationships
5. **Activity Analysis** - behavioral patterns (posting frequency, timing, coordination)
6. **Content Analysis** - thematic sub-sections organized by narrative cluster
   - e.g., "Racism and Xenophobia," "Conspiracy Communities," "Health Misinformation," "Geopolitical Tensions"
7. **Cross-Campaign Overlap** - comparative analysis showing actor reuse across operations
8. **Conclusions** - synthesis of findings

This structure reflects Graphika's core competency: they analyze the **network** first, then the content within that network context.

### Attribution Model
- **Network-based attribution**: Attributes behavior to clusters/communities, not just individual accounts
- Analyzes billions of interactions to identify coordinated behavior at scale
- Cross-campaign tracking — showed 40%+ account overlap between White Helmets disinfo and other operations, with 92.6% from the same network cluster
- Uses their proprietary **Coordination Framework** technology
- Partners with platforms (Meta, Pinterest) for dataset access

### Confidence Signaling
- **Multi-dimensional confidence**: Analyzes anomalies across three dimensions:
  - **Network** anomalies (unusual connection patterns)
  - **Semantic** anomalies (content patterns)
  - **Temporal** anomalies (timing/coordination)
- Confidence implicitly scales with how many dimensions show anomalies
- Language: "analysis revealed," "patterns consistent with," "indicators suggest"
- Quantitative metrics provide implicit confidence (e.g., "92.6% overlap")

### Evidence Presentation
- **Network maps** — large-scale visualizations showing community structures (their signature output)
- Quantitative metrics: millions of posts analyzed, overlap percentages, engagement rates
- Temporal analysis charts showing activity spikes correlated with events
- Content clustering analysis
- Cross-campaign comparison data
- Case-specific deep dives within the network context

### Visual Elements
- **Network maps are the primary visual product** — Graphika is "the cartographers of the internet age"
- Large-scale node-link diagrams with color-coded communities
- Temporal heatmaps showing activity patterns
- Comparative network snapshots across time periods
- Thematic content clustering visualizations
- Geographic distribution maps

### Actionability
- Highly actionable for platforms — identifies specific networks for takedown
- Actionable for government/intelligence — identifies state-linked operations
- Cross-campaign tracking enables proactive detection of future operations
- Network maps provide strategic overview for decision-makers

### Strengths
- Unique network-analysis-first methodology
- Scale: analyzes billions of interactions
- Multi-dimensional analysis (network + semantic + temporal)
- Cross-campaign tracking reveals persistent threat actors
- Proprietary technology (Contagion Monitor, Coordination Framework)
- Academic partnerships (Cornell, Johns Hopkins APL via MINERVA grant)
- Multilingual and multi-platform analysis

### Weaknesses
- Proprietary methodology limits reproducibility
- Network maps require expertise to interpret
- Focus on large-scale campaigns — less suited to individual account analysis
- Dependency on platform data access
- Reports can be technically dense

### CDDBS Takeaways
- **Adopt**: Multi-dimensional analysis framework (network + content + temporal)
- **Adopt**: Cross-campaign/cross-account overlap tracking
- **Adopt**: Quantitative metrics as evidence (percentages, counts, rates)
- **Adopt**: Temporal analysis (activity patterns over time)
- **Learn from**: Network visualization approach — even at individual account level, show connections
- **Improve on**: Make methodology transparent and reproducible (unlike proprietary approach)
- **Improve on**: Adapt network analysis for individual account level (CDDBS's primary use case)

## Briefing 7: RAND Corporation

### Source
- **Organization**: RAND Corporation
- **Title**: "The Russian 'Firehose of Falsehood' Propaganda Model" (PE-198-OSD) by Christopher Paul & Miriam Matthews
- **Date**: 2016 (seminal report); follow-up reports on production/distribution/consumption
- **URL**: https://www.rand.org/pubs/perspectives/PE198.html

### Structure
RAND "Perspectives" reports follow a compact policy brief format:
1. **Title Page** - with RAND branding and sponsoring organization (OSD in this case)
2. **Key Findings Box** - highlighted summary of core conclusions (immediately visible)
3. **Introduction** - framing the problem
4. **Model Description** - defining the phenomenon (the four characteristics of the firehose)
5. **Evidence / Why It Works** - citing psychological research supporting each characteristic
6. **Implications** - what this means for defenders
7. **Recommendations / Options to Counter** - specific countermeasures
8. **References** - academic bibliography

Follow-up RAND reports expanded the framework into three phases:
- **Production** — how content is created (malinformation, misinformation, disinformation)
- **Distribution** — how content spreads via social networks
- **Consumption** — how content is received by the public

### Attribution Model
- **Academic citation** with numbered endnotes and full bibliography
- References to psychological research literature (peer-reviewed journals)
- Cites specific examples from Russian media operations with dates and outlets
- Attributes the propaganda model to observed behaviors rather than to intelligence sources
- Draws exclusively on publicly available sources (deliberate choice for credibility)

### Confidence Signaling
- **Explicitly cautious**: "not optimistic about the effectiveness of traditional counterpropaganda efforts"
- Uses conditional language: "may be possible," "evidence suggests," "research indicates"
- Distinguishes between the model description (high confidence — observed behavior) and counter-strategies (moderate confidence — proposed but unproven)
- Cites specific psychological studies to ground confidence in behavioral claims
- The four characteristics are presented as **observed and documented**, not speculative

### Evidence Presentation
- **Literature review** — draws on existing psychological and communications research
- Specific examples from Russian media operations (RT, Sputnik, social media)
- References to academic experiments on repetition, source credibility, and misinformation persistence
- Compact format — evidence is woven into the narrative, not in separate sections
- The "firehose" metaphor itself is a powerful framing device

### Visual Elements
- Minimal — RAND Perspectives are text-heavy policy briefs
- Key findings box is visually highlighted
- Clean, professional layout with RAND branding
- No network graphs, charts, or images
- Brevity is the visual strategy — the entire report is ~11 pages

### Actionability
- **Highly actionable** — concludes with specific counter-strategies:
  1. "Pre-bunking" (inoculation) over debunking
  2. Focus on effects, not the propaganda itself
  3. "Compete" — redirect audiences rather than engage with the firehose
  4. Teach digital literacy
- Recommendations are framed as options, not mandates
- Directly addresses policymakers and defense professionals

### Strengths
- Concise and accessible — 11 pages, clear language
- Powerful conceptual framework (the "firehose" model is now widely cited)
- Grounded in psychological research (not just observation)
- Actionable recommendations with clear rationale
- Publicly available — builds credibility through transparency
- Influential — has shaped NATO and allied counter-disinformation thinking

### Weaknesses
- High-level/strategic — doesn't provide tactical detection tools
- No quantitative data analysis of specific campaigns
- No visual evidence or screenshots
- Published in 2016 — some aspects may need updating for AI-era disinformation
- Counter-strategies are proposals, not tested interventions

### CDDBS Takeaways
- **Adopt**: Key findings box prominently displayed at top of briefing
- **Adopt**: Grounding assessments in research literature, not just observation
- **Adopt**: Conciseness — prioritize clarity over comprehensiveness
- **Adopt**: Counter-strategy recommendations (what can be done about this account/campaign?)
- **Learn from**: The "firehose" framework — categorize disinformation by production/distribution/consumption
- **Learn from**: Making the distinction between what is observed and what is recommended
- **Improve on**: Add quantitative data analysis and visual evidence (which RAND doesn't include)

## Briefing 8: UK DCMS Committee

### Source
- **Organization**: UK House of Commons Digital, Culture, Media & Sport Committee
- **Title**: "Disinformation and 'fake news': Final Report" (HC 1791)
- **Date**: February 14, 2019 (after 18-month inquiry started September 2017)
- **URL**: https://publications.parliament.uk/pa/cm201719/cmselect/cmcumeds/1791/179102.htm

### Structure
The DCMS report follows a **parliamentary select committee** format:
1. **Title and Formal Header** - committee membership, terms of reference
2. **Summary** - condensed findings and key recommendations
3. **Introduction and Background** - scope of inquiry, methodology
4. **Thematic Chapters** (7 main issue areas):
   - Definition and role of tech companies
   - Electoral law and foreign interference
   - Data protection and privacy
   - Regulation and enforcement
   - Digital literacy and education
   - Advertising and political campaigning
   - Platform accountability
5. **Numbered Recommendations** — each tied to a specific finding
6. **Conclusions** - overall assessment
7. **Formal Minutes** - procedural record
8. **Witnesses** - list of oral evidence sessions
9. **Published Written Evidence** - submissions from organizations and individuals

Scale: 23 oral evidence sessions (including Washington DC), 170+ written submissions, 73 witnesses, with an international "Grand Committee" (first since 1933) involving representatives from 8 countries.

### Attribution Model
- **Witness testimony**: Direct quotes attributed to named witnesses with session dates
- **Written submissions**: Referenced by author/organization and submission number
- **Cross-referencing**: Links between evidence from different witnesses
- Formal parliamentary privilege allows naming of individuals and organizations
- The **Interim Report** (July 2018) is referenced and built upon

### Confidence Signaling
- **Parliamentary authority language**: "The Committee believes," "We recommend," "We are concerned"
- Confidence is institutional — the committee speaks as one voice
- Uses strength of language to signal conviction: "We are deeply concerned" vs. "We note"
- Recommendations carry implicit weight of parliamentary authority
- Dissenting views, when present, are formally recorded

### Evidence Presentation
- **Oral testimony transcripts** — verbatim quotes from hearings
- **Written submissions** — formal evidence from tech companies, civil society, academics
- Cross-examination format reveals contradictions (e.g., Facebook's evolving testimony)
- Document analysis of internal company communications
- Reference to academic research and international reports
- No quantitative social media data analysis

### Visual Elements
- Minimal — standard parliamentary report format (text-heavy)
- Numbered paragraphs for precise referencing

### Key Insight: The Confidence Gap
**Only 3 of 10 organizations use explicit confidence signaling.** Most rely on hedging language ("suggests," "consistent with," "indicates") without a formal scale. This is a major opportunity for CDDBS — adopting a formal confidence framework would differentiate our briefings from most existing products.

### Attribution Best Practices

**Tier 1 — Gold Standard** (Bellingcat, DFRLab):
- Every claim traced to specific evidence (screenshot, URL, archived post)
- Tools used for verification explicitly named
- Evidence chains documented to legal admissibility standards (Bellingcat J&A Unit)
- Distinction between "verified" (independently confirmed) and "assessed" (analytical judgment)

**Tier 2 — Strong** (Graphika, SIO, OII):
- Quantitative evidence cited with methodology
- Platform-provided datasets with documented selection criteria
- Cross-organizational corroboration
- Academic citation standards

**Tier 3 — Adequate** (EUvsDisinfo, NATO StratCom, GEC, RAND, UK DCMS):
- Sources named and linked
- Claims supported by references
- Less granular tracing to individual data points
- Relies more on institutional credibility than evidence chains

**CDDBS Target**: Tier 2 with elements of Tier 1. Every finding should trace to specific evidence (posts, metadata, behavioral patterns), with methodology disclosed. Tier 1 legal-grade documentation is aspirational but beyond current scope.

## CDDBS Template Proposal

Based on the analysis of 10 organizations above, here is the proposed CDDBS intelligence briefing format for v1.1:

### Proposed Structure

The CDDBS briefing should have **7 mandatory sections** in this order:

| # | Section | Inspired By | Purpose |
|---|---------|-------------|---------|
| 1 | **Executive Summary** | RAND (key findings box), NATO StratCom | 2-3 sentences: what was found, why it matters, confidence level |
| 2 | **Key Findings** | All organizations | 3-5 numbered findings, each with confidence level and evidence reference |
| 3 | **Subject Profile** | EUvsDisinfo (metadata), DFRLab | Structured account metadata (handle, platform, dates, counts) |
| 4 | **Narrative Analysis** | GEC (pillars), OII (framework), Graphika (multi-dimensional) | Primary narratives, behavioral indicators, network context, source attribution |
| 5 | **Confidence Assessment** | Graphika (multi-dimensional), Bellingcat (explicit gaps) | Overall + per-factor ratings, limitations, alternative explanations |
| 6 | **Limitations & Caveats** | Bellingcat, SIO | What the analysis cannot determine, known gaps |
| 7 | **Methodology** | SIO, OII, DFRLab | Data collection, analysis model, validation steps |

### Proposed Attribution Model

Drawing from Bellingcat (Tier 1) and Graphika (Tier 2):

1. **Every finding must reference specific evidence** — no unattributed claims
2. **Evidence types must be labeled**:
   - `[POST]` — specific social media post (with URL or archived reference)
   - `[PATTERN]` — observed behavioral pattern (with metrics)
   - `[NETWORK]` — network/relationship data
   - `[METADATA]` — account metadata (creation date, bio changes, etc.)
   - `[EXTERNAL]` — reference to external source or database
3. **Attribution language must be precise** (from Bellingcat/DFRLab):
   - "The account posted..." (observed fact)
   - "This is consistent with..." (pattern matching)
   - "We assess with [level] confidence that..." (analytical judgment)
   - "This suggests..." (inference)

### Proposed Confidence Scale

Synthesized from NATO StratCom, Graphika, and intelligence community standards:

**Three-level scale** (matching Graphika's dimensional approach):

| Level | Definition | Requirements |
|-------|-----------|-------------|
| **High** | Multiple independent indicators consistent with assessment; strong corroboration across content, behavioral, and network analysis | 3+ indicator dimensions align; no significant contradicting evidence |
| **Moderate** | Assessment supported by available evidence but with gaps; limited corroboration or some conflicting indicators | 1-2 indicator dimensions align; some gaps or ambiguity |
| **Low** | Plausible assessment based on limited evidence; alternative explanations remain viable | Single indicator dimension; significant gaps; viable alternatives exist |

**Per-finding confidence** (CDDBS innovation — not found in any of the 10 organizations):
- Each key finding gets its own confidence rating
- Overall briefing confidence is the lowest per-finding confidence

**Confidence factors** (inspired by Graphika's three-dimensional analysis):
- Data completeness — how complete is the available data?
- Source reliability — how reliable are the data sources?
- Analytical consistency — how consistent are the indicators?
- Corroboration — multiple independent indicators?

### Proposed Visual Elements (v1.1 Priority)

Based on the visual analysis, prioritized by implementation feasibility:

**v1.1 (Sprint 1)**:
- Key findings box with color-coded confidence badges (High=green, Moderate=yellow, Low=red)
- Subject profile metadata table
- Confidence assessment table with factor breakdown

**v1.2 (Sprint 2)**:
- Activity timeline chart (posting frequency over time)
- Narrative frequency breakdown (bar chart)

**v1.3+ (Future)**:
- Network graph (connections to other analyzed accounts)
- Annotated tweet screenshots
- Geographic distribution map

### Key Design Decisions

| Decision | Choice | Rationale |
|----------|--------|-----------|
| Confidence scale | 3-level (H/M/L) | Industry standard (intelligence community), simple enough for LLM output |
| Per-finding confidence | Yes | No organization does this — CDDBS innovation that adds analytical precision |
| Mandatory limitations | Yes | Bellingcat and SIO demonstrate this builds trust |
| Methodology section | Yes | 8/10 organizations include this; critical for credibility |
| JSON + Human-readable | Both | JSON for programmatic use, rendered HTML/Markdown for analysts |
| Attribution language | Standardized vocabulary | Four phrases covering observation → inference → assessment |
| Account-level focus | Yes | CDDBS analyzes individual accounts (unlike Graphika's network-level) |
| Ecosystem positioning | Database + Policy Brief hybrid | Scalable and consistent (like EUvsDisinfo) but analytically deep (like RAND) |
