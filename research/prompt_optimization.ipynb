### Recommended Architecture for v1.2

Based on research findings, the recommended prompt architecture:

1. **Role definition** (fixed) — intelligence analyst persona
2. **Known narratives** (updateable) — reference dataset injected from `known_narratives.json`
3. **Output format** (fixed) — 7 mandatory sections with field requirements
4. **Attribution rules** (fixed) — 4 patterns + NEVER list
5. **Confidence calibration** (fixed) — dimensional requirements + base rate awareness
6. **Self-validation** (fixed) — pre-output checklist
7. **JSON mode instructions** (conditional) — only when JSON output requested
